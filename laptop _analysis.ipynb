##### helper function to extract numerical values

```python
def extract_int(text):
    if pd.isna(text):return np.nan
    s=str(text)
    m=re.search(r'(\d+)',s)
    return int(m.group(1))if m else np.nan
```

```python
def extract_storage_gb(text):
    if pd.isna(text):return 0
    s=str(text).lower()
    total=0
    for m in re.finditer(r'(\d+)\s*gb',s):
        total+=int(m.group(1))
    for m in re.finditer(r'(\d+)\s*tb',s):
        total+=int(m.group(1))*1000
    if total==0:
         m=re.search(r'(\d+)',s)
         total=int(m.group(1))if m else 0
    return total
```

```python
def extract_ghz(text):
    if pd.isna(text):return np.nan
    s=str(text)
    m=re.search(r'([\d\.]+)\s*ghz',s,flags=re.IGNORECASE)
    if m:
        return float(m.group(1))
        m2=re.search(r'[\d\.]+)',s)
        return float(m2.group(1)) if m2 else np.nan
```

```python
def extract_battery_hours(text):
    if pd.isna(text): return np.nan
    s=str(text).lower()
    m=re.search(r'([\d\.]+)\s*(hours/hrs/hr/h)\b',s)
    if m: return float(m.group(1))
    m2=re.search(r'([\d\.]+)',s)
    return float(m2.group(1)) if m2 else np.nan
```

```python
def extractModel(x):
    lb=str(x).split()
    model="".join(lb[1:2])
    return model
```

#####  extract numeric feature

```python
df['RAM_GB']=df['RAM'].astype(str).apply(extract_int).astype(int)
df['SSD_GB']=df['SSD'].astype(str).apply(extract_storage_gb).astype(int)
df['HDD_GB']=df['HDD'].astype(str).apply(extract_storage_gb).astype(int)
df['Total_Storage_GB']=df['SSD_GB']+df['HDD_GB']
df['Ghz_num']=df['Ghz'].astype(str).apply(extract_ghz).astype(float)
df['Battery_hours']=df['Battery_Life'].astype(str).apply(extract_battery_hours).astype(float)
df['GPU_Brand']=df['GPU_Brand'].fillna('Unknown')
df['Has_SSD']=df['SSD_GB']>0
df['Model']=data['Name'].apply(extractModel)
```
